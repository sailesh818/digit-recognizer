```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python

img = cv2.imread('digit_four.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img_resized = cv2.resize(img, (600, 100))
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_resized, cmap='gray')
plt.title("Resized Image")
plt.axis('off')
plt.show()
```

```python
_, image_thresh= cv2.threshold(img_resized, 128, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(image_thresh, cmap="gray")
plt.title("Image with handwritten multiple letters")
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

flattened_image = image_thresh.flatten().reshape(1, -1)

scaler = StandardScaler()
flattened_scaled = scaler.fit_transform(flattened_image)

```

```python
resized_image = cv2.resize(image_thresh, (28, 28))
flattened_image = resized_image.flatten().reshape(1, -1)
X_train_dummy = np.random.randint(0, 255, (5, 784))
y_train_dummy = np.array([0, 1, 2, 3, 4])

scaler_dummy = StandardScaler()
X_train_scaled = scaler_dummy.fit_transform(X_train_dummy)
model = SVC(kernel='rbf', gamma='auto', C=1)
model.fit(X_train_scaled, y_train_dummy)

flattened_scaled = scaler_dummy.transform(flattened_image)
prediction = model.predict(flattened_scaled)
print("Predicted label:", prediction[0])

plt.imshow(resized_image, cmap='gray')
plt.title(f"Predicted: {prediction[0]}")
plt.axis('off')
plt.show()
```
